```
def lcm2(a, b):
    import math
    return a * b // math.gcd(a, b)
### ADVANCED
### functional lcm :: Using reduce
from functools import reduce
def lcmf(many):
    return reduce(lcm2, many)
lcmf(range(1, 10))
```
