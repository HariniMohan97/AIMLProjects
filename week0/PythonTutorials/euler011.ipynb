```python
grid = np.array([int(x) for x in data.split()])
```

```python
def hprod(row):
    return [np.prod(grid[row, col:col+4]) for col in range(0, 17)]
```

```python
def rowmax():
    for row in range(17):
        return max([max(hprod(row))])
      
def v_rowmax():
  return max([max(hprod(row)) for row in range(17)])
```
